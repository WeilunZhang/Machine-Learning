```python
from selenium import webdriver
driver = webdriver.Firefox()
```

```python
driver.get("http://search.sina.com.cn/")
driver.find_element_by_xpath('//*[@id="tabc02"]/form/div/input[1]').send_keys("百度")
```

```python
driver.find_element_by_xpath('//*[@id="tabc02"]/form/div/input[4]').click()
results = driver.find_elements_by_xpath('//*[@id="result"]')
```

```python
for result in results:
    print(result.text)
```
